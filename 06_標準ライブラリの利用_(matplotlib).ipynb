##【課題1】次の３次関数のグラフを描いてください。(注) 関数に加えて，x の範囲(＝定義域)も変更されている。

$y=2x^3-3x^2-4x+5 \quad (-5 \le x \le 5)$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-5, 5, 0.1)

y = 2 * x**3 - 3 * x**2 - 4 * x + 5

plt.plot(x, y)
plt.show()
```

（解説）数学における数式を Python 言語へ落とし込めば OK です。

なお，「xの２乗」や「xの３乗」に power() を使いたい場合， `math.pow()` でなく，`np.power()` を使ってください。

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-5, 5, 0.1)

y = 2 * np.power(x,3) - 3 * np.power(x,2) - 4 * x + 5

plt.plot(x, y)
plt.show()
```

## 【課題2】次の三角関数を含む関数のグラフを描いてください。
$y = 3 \sin(x) + \sin(3x) \quad (-\pi \le x \le \pi)$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-np.pi, np.pi, 0.1)

y = 3 * np.sin(x) + np.sin(3*x)

plt.plot(x, y)
plt.show()
```

(解説) sin 関数は `np.sin()` で与えられることに注意。

## 【課題3】次の関数のグラフを描いてください。

$z = x^2 - y^2 \quad (-5 \le x \le 5, -5 \le y \le 5)$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.arange(-5.0, 5.0, 0.1)
y = np.arange(-5.0, 5.0, 0.1)
X, Y = np.meshgrid(x, y)
Z = X**2 - Y**2

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_wireframe(X, Y, Z)
plt.show()
```

(解説) 上記のグラフは「馬鞍形」と呼ばれる。

## 【課題4】次の関数のグラフを描いてください。

$z = \sin(x) \cdot \sin(y) (-5 \le x \le 5, -5 \le y \le 5)$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.arange(-5.0, 5.0, 0.1)
y = np.arange(-5.0, 5.0, 0.1)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.sin(Y)

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_wireframe(X, Y, Z)
plt.show()
```

（解説）それなりに複雑な形のグラフであり，最適化アルゴリズムを学ぶ際に頻繁に出てきます。

### (余談) 最適化に絡み，Rastrigin 関数を紹介します。  
むちゃギザギザしています。

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.arange(-5.12, 5.12, 0.1)
y = np.arange(-5.12, 5.12, 0.1)
X, Y = np.meshgrid(x, y)
Z = 10 + (x**2 - 10 * np.cos(2 * np.pi * X)) + 10 + (Y**2 - 10 * np.cos(2 * np.pi * Y)) # ２次元の Rastrigin 関数

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X, Y, Z)
plt.show()
```
